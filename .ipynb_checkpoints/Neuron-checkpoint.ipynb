```python
import torch
import torch.nn as nn #It includes buildings blocks such as activation function and layer creation 
```

```python
#input
torch.manual_seed(42)
inputs = torch.tensor([1.0, 2.0, 3.0]) #(x1, x2, x3)
```

```python
neuron = nn.Linear(in_features=3, out_features=1) #linear activation function for weight calculation
```

```python
output = neuron(inputs)
```

```python
output
```

```python
# verify using formula
#bias value initialize randomly so every time output will change without "torch.manual_seed(42)"
neuron.weight @ inputs + neuron.bias # @ --> for dot product
```
